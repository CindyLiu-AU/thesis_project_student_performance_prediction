### filter course BBB

```python
df_studentVle_BBB = df_studentVle_ori[df_studentVle_ori.code_module == 'BBB']
df_studentVle_BBB
```

```python
df_studentInfo_BBB = df_studentInfo_ori[df_studentInfo_ori.code_module == 'BBB']
df_studentInfo_BBB
```

### create unique id for both datasets

```python
df_studentInfo_BBB['unique_id'] = df_studentInfo_BBB['code_module'] + '_' + df_studentInfo_BBB['code_presentation'] + '_' + df_studentInfo_BBB['id_student'].apply(str)
df_studentInfo_BBB.drop(['code_module','code_presentation', 'id_student'], axis=1, inplace= True)
df_studentInfo_BBB
```

```python
df_studentVle_BBB['unique_id'] = df_studentVle_BBB['code_module'] + '_' + df_studentVle_BBB['code_presentation'] + '_' + df_studentVle_BBB['id_student'].apply(str)
df_studentVle_BBB.drop(['code_module','code_presentation', 'id_student'], axis=1, inplace= True)
df_studentVle_BBB
```

## prepare data based on the research purpose
### (1) only use non-withdrawn student (2) make two labels - fail / pass

```python
df_studentInfo_BBB = df_studentInfo_BBB[df_studentInfo_BBB['final_result'] != 'Withdrawn']
df_studentInfo_BBB
```

```python
df_studentInfo_BBB.loc[(df_studentInfo_BBB.final_result == 'Distinction'), 'final_result'] = 'Pass'
```

```python
df_studentInfo_BBB['final_result'].unique()
```

```python
df_studentInfo_BBB.groupby(by='final_result').size()
```

```python
df_studentInfo_BBB.to_csv('temp_data_studentInfo.csv',index=False) 
```

# prepare student click behaviour data

```python
df_vle_ori = pd.read_csv('raw_data_vle.csv', low_memory = False)
df_vle_BBB = df_vle_ori[df_vle_ori['code_module'] == 'BBB']
df_vle_BBB
```

```python
df_merged = pd.merge(df_studentVle_BBB, df_vle_BBB, how='left', left_on = 'id_site', right_on = 'id_site')
df_merged
```

```python
df_merged.drop(['id_site','code_module', 'code_presentation','week_from', 'week_to'], axis=1, inplace= True)
```

```python
df_merged.nunique() 
```

```python
df_click_behaviour = df_merged.pivot_table(df_merged, index=['unique_id','activity_type', 'date'], aggfunc='sum')
df_click_behaviour
```

```python
temp = df_click_behaviour.reset_index(level='date').reset_index(level='activity_type').reset_index(level='unique_id')
temp
```

```python
temp.nunique() 
# In this temp data of students' click behaviour, 6622 students are recorded. 
# However, in studentInfo data, only 5521 students are recorded. 
```

```python
temp.to_csv('temp_data_click_behaviour.csv',index=False) 
temp
```
