```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
import math
```

# load data - studentVle

### load studentVle.csv

```python
temp_data_studentInfo = pd.read_csv('temp_data_studentInfo.csv', low_memory = False)
temp_data_studentInfo
```

```python
temp_data_studentInfo = temp_data_studentInfo.iloc[:, -2:]
temp_data_studentInfo
```

```python
click_behaviour = pd.read_csv('temp_data_click_behaviour.csv', low_memory = False)
click_behaviour
```

```python
click_behaviour.info()
```

# strategy1 data
### click behaviours are observed by time period (T)

```python
df_click_by_date = click_behaviour.pivot_table(click_behaviour, index=['unique_id','date'], aggfunc='sum')
df_click_by_date
```

```python
df_click_by_date_reset_index = df_click_by_date.reset_index(level='date').reset_index(level='unique_id')
df_click_by_date_reset_index
```

```python
df_click_by_date_pivoted = df_click_by_date_reset_index.pivot(index='unique_id', columns='date', values='sum_click').reset_index().fillna(0)
df_click_by_date_pivoted
```

# T0

```python
df_early_access = df_click_by_date_pivoted.loc[:,:0]
```

```python
df_early_access['T0'] = df_early_access.sum(axis=1)
df_early_access
```

```python
df_early_access = df_early_access[['unique_id','T0']]
df_early_access
```

# T1...Tn for weekly view (WEE)

```python
df1 = df_click_by_date_pivoted.loc[:,1:]
df2 = df_click_by_date_pivoted.loc[:,1:]
```

```python
total_days = list(range(1, 269))
print(total_days)
```

```python
days_in_a_week = 7
for i in range(math.ceil(269/days_in_a_week)):
    col_names_week = 'T' + str(i+1)
#     print(col_names_week)
    aggre_index_week = total_days[days_in_a_week*i:days_in_a_week*(i+1)]
#     print(aggre_index_week)
    #define new column that contains sum of specific columns
    df1[col_names_week] = df1[aggre_index_week].sum(axis=1)
df1
```

```python
df_course_start_WEE = df1.loc[:,'T1':]
df_course_start_WEE
```

# T0 + T1..Tn for WEE

```python
S1WEE = df_early_access.join(df_course_start_WEE)
S1WEE
```

```python
df_S1WEE = pd.merge(temp_data_studentInfo, S1WEE, how='inner', left_on="unique_id", right_on="unique_id")
df_S1WEE
```

```python
df_S1WEE.columns
```

# T1...Tn for monthly view (MON)

```python
days_in_a_month = 30
for j in range(math.ceil(269/days_in_a_month)):
    col_names_month = 'T' + str(j+1)
#     print(col_names_month)
    aggre_index_month = total_days[days_in_a_month*j:days_in_a_month*(j+1)]
#     print(aggre_index_month)
    #define new column that contains sum of specific columns
    df2[col_names_month] = df2[aggre_index_month].sum(axis=1)
df2
```

```python
df_course_start_MON = df2.loc[:,'T1':]
df_course_start_MON
```

# T0 + T1..Tn for MON

```python
S1MON = df_early_access.join(df_course_start_MON)
S1MON
```

```python
df_S1MON = pd.merge(temp_data_studentInfo, S1MON, how='inner', left_on="unique_id", right_on="unique_id")
df_S1MON
```

```python
df_S1MON.columns
```
